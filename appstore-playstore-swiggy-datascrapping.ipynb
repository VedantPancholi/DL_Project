```python
from google_play_scraper import reviews_all, Sort
from app_store_scraper import AppStore
import pandas as pd

# üéØ Google Play Store Scraper
app_package = "in.swiggy.android"  # Google Play package name

g_reviews = reviews_all(
    app_package,
    sleep_milliseconds=0,  
    lang="en",             
    country="in",          
    sort=Sort.NEWEST       
)

# Convert to DataFrame
df_google = pd.DataFrame(g_reviews)

# ‚úÖ Add Extra Columns
df_google["Company"] = "Swiggy"
df_google["Platform"] = "Google Play Store"


```

```python
# üéØ Apple App Store Scraper (Corrected ID)
app_store_id = "989540920"  # Corrected Apple App Store ID

a_reviews = AppStore(country="in", app_name="swiggy-food-instamart-dineout", app_id=app_store_id)
a_reviews.review()  

# Convert to DataFrame
df_apple = pd.DataFrame(a_reviews.reviews)

# ‚úÖ Add Extra Columns
df_apple["Company"] = "Swiggy"
df_apple["Platform"] = "App Store"

# üéØ Combine Both DataFrames
df_combined = pd.concat([df_google, df_apple], ignore_index=True)

# Print the combined DataFrame
print("‚úÖ Combined Reviews:\n", df_combined.head())

# ‚úÖ Display Shape of Data
print(f"\nTotal Reviews Combined: {df_combined.shape[0]}")
```

```python
# ‚úÖ Create a copy to avoid SettingWithCopyWarning
df4_apple = df3_apple.copy()

# ‚úÖ Extract 'body' from 'developerResponse' safely
df4_apple["reply"] = df4_apple['developerResponse'].apply(lambda x: x.get("body") if isinstance(x, dict) else None)

# ‚úÖ Drop the 'developerResponse' column correctly
df4_apple.drop(columns=["developerResponse"], inplace=True)

df4_apple.reset_index()
```
